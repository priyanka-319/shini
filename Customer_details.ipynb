```python
import pandas as pd
import itertools

customer_data_one = pd.read_csv('Downloads/customer_data_one.csv')
customer_data_two = pd.read_csv('Downloads/customer_data_two.csv')
```

```python
customer_data_one.head()
```

```python
customer_data_two.head()
```

```python
nos_in_customer_data_two = [x for x in customer_data_two.mobile_number]
```

```python
for _,row in customer_data_one.iterrows():
    if row['mobile_number'] in nos_in_customer_data_two:
            #print(row['mobile_number'])
            customer_data_one.drop(index=_,inplace=True)
        
```

```python
result = pd.concat([customer_data_two,customer_data_one],ignore_index=True)
result.set_index('mobile_number',inplace=True)
result.sort_values('first_name')
```

```python
result.shape
```

```python
result.to_json('Downloads/json_file.json',orient = 'index')
```
